```python
GOAL="GOAL1"

path= "/home/notebook/pkl/results/GOAL1/all_objects/"
```

```python
import json, os
from glob import glob
```

```python
if (GOAL=="GOAL1"):
    results_file_ap= 'ap_results.json'
    results_file_PKL_CRIT= 'pkl_crit_results_GOAL1.json'
    results_file_PKL= 'pkl_results.json'
if(GOAL=="GOAL2"):
    results_file_PKL_CRIT= 'pkl_crit_results_GOAL2.json'
    results_file_ap="None"
    results_file_PKL="None"
```

```python
directories=glob(path+"/*/", recursive = True)
```

```python
def analyze_PKL_CRIT(path, file_name):
    f = open(path+"/"+file_name)
    # returns JSON object as 
    # a dictionary
    data = json.load(f)
    results=[]
    for i in data['PKL_CRIT_RESULTS']:
        for d in i:
            results.append([d['min'],
                           d['max'],
                           d['mean'],
                           d['median'],
                           d['confidence'],
                           d['crit'],
                           d['D'],
                           d['R'],
                           d['T'],
                          ])


    return results


def analyze_PKL_CRIT_GOAL2(path, file_name):
    f = open(path+"/"+file_name)
    # returns JSON object as 
    # a dictionary
    data = json.load(f)
    results=[]
    for i in data['PKL_CRIT_RESULTS_GOAL2']:
        for d in i:
            results.append([d['min'],
                           d['max'],
                           d['mean'],
                           d['median'],
                           d['confidence'],
                           d['crit'],
                           d['D'],
                           d['R'],
                           d['T'],
                          ])


    return results
```

```python
def analyze_PKL(path, file_name):
    f = open(path+"/"+file_name)
    # returns JSON object as 
    # a dictionary
    data = json.load(f)
    results=[]
    for i in data['PKL_RESULTS']:
        for d in i:
            results.append([d['min'],
                           d['max'],
                           d['mean'],
                           d['median'],
                           d['confidence'],
                          ])
    return results
```

```python
def analyze_AP(path, file_name):
    f = open(path+"/"+file_name)
    # returns JSON object as 
    # a dictionary
    data = json.load(f)
    results=[]
    for i in data['AP_RESULTS']:
        results.append([i['label_aps'],
                        i['label_aps_crit'],
                        i['DRT'],
                        i["mean_ap"],
                        i["mean_ap_crit"], 
                       ])

    return results
```

```python
results_total={}
if (GOAL=="GOAL1"):
    for d in directories:
        directory_name=d
        files=os.listdir(d)
        results_PKL_CRIT=[]
        results_PKL=[]
        results_AP=[]
        for f in files:
            if(f==results_file_PKL_CRIT):
                results_PKL_CRIT=analyze_PKL_CRIT(d, results_file_PKL_CRIT)
            if(f==results_file_PKL):
                results_PKL=analyze_PKL(d,results_file_PKL)
            if(f==results_file_ap):
                results_AP=analyze_AP(d,results_file_ap)
        results_total.update({str(directory_name): [results_PKL, results_PKL_CRIT, results_AP]})
if(GOAL=="GOAL2"):
    for d in directories:
        directory_name=d
        files=os.listdir(d)
        results_PKL_CRIT=[]
        results_PKL=[]
        results_AP=[]
        for f in files:
            if(f==results_file_PKL_CRIT):
                results_PKL_CRIT=analyze_PKL_CRIT_GOAL2(d, results_file_PKL_CRIT)
            if(f==results_file_PKL):
                results_PKL=analyze_PKL(d,results_file_PKL)
            if(f==results_file_ap):
                results_AP=analyze_AP(d,results_file_ap)
        results_total.update({str(directory_name): [results_PKL, results_PKL_CRIT, results_AP]})
    
```
